# Administering basic attacks and visualizing/plotting their effects in various configurations

## Sentence Level Positive Word Injection attack
 We simply select words that occur multiple times
in the training data in benign comments, but relatively fewer times in negative training comments and add these words to the start and end of a sentence
based on a percentage of the sentence length. This percentage is determined by the attack strength. Data needs to be preprocessed after running
this attack

```python
from positive_words_sentence_attack import pos_words_sent_attack
import os
import sys
import inspect
import predict_utils as pred
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python

data_folder = r'preprocessed/try_2/'
attack_name = "adv_pos_words_sent"
attack_folder = os.path.join(data_folder,attack_name)
```

### Eval on baseline model

```python
top_n = 50
scores = []
BEST_THRESHOLD = 0.3
for attack_strength in range(0,11):
    attack_strength = round(attack_strength*0.1,2)
    data_folder = r'preprocessed/try_2/'
    attack_name = "adv_pos_words_sent"
    # path to save adversarial examples
    adv_folder = os.path.join(data_folder, f"{attack_name}_{str(attack_strength).replace('.', '_')}_{top_n}")
    
    

    
    confidences, predictions, ground_truth, alphas = pred.get_predictions(r'model_new', adv_folder, data_type="", dump_folder_path="dump_2",
                    model_type="baseline")
        
    tn, fp, fn, tp = confusion_matrix(ground_truth, [1 if y > BEST_THRESHOLD else 0 for y in predictions]).ravel()
    # attack success is the number of false negatives
    attack_sucess = fn
    print(f"++++++++++++++++{attack_strength}+++++++++++++++++++++")
    roc_auc,pr_auc,f1_score,accuracy = pred.get_metrics(ground_truth,confidences)
    pred.plot_confusion_matrix(ground_truth, predictions, f"CONFUSION MATRIX: BASELINE SENT POS WORDS: {attack_strength}")
    pred.plot_confusion_matrix(ground_truth, predictions, f"ROC AUC: BASELINE SENT POS WORDS: {attack_strength}")
    scores.append({"roc_auc":roc_auc, "pr_auc":pr_auc, "f1": f1_score, "accuracy":accuracy, "attack_strength":attack_strength,
                  "attack_success":fn})
    
    
    
    
    
    
baseline_scores_df = pd.DataFrame(scores)    
```

```python
baseline_scores_df = pd.DataFrame(scores)
```

```python
import matplotlib.pyplot as plt


with plt.style.context('Solarize_Light2'):
    plt.plot(baseline_scores_df.attack_strength, baseline_scores_df.accuracy, label="accuracy")
    plt.plot(baseline_scores_df.attack_strength, baseline_scores_df.f1, label="f1_score")
    plt.plot(baseline_scores_df.attack_strength, baseline_scores_df.roc_auc, label="roc_auc")
    plt.plot(baseline_scores_df.attack_strength, baseline_scores_df.pr_auc, label="pr_auc")
    leg = plt.legend(loc= 'center')
    plt.title("Sentence Level Positive word Injection Attack on Baseline model")
    plt.xlabel('attack strength (% sentence length word insertions)')
    plt.ylabel('Model Performance' )
plt.show()
```

```python
import matplotlib.pyplot as plt


with plt.style.context('Solarize_Light2'):
    plt.plot(baseline_scores_df.attack_strength, baseline_scores_df.attack_success, label="attack success (# of false negatives)")
    plt.title("Sentence Level Positive word Injection Attack on Baseline model")
    plt.xlabel('attack strength (% sentence length word insertions)')
    plt.ylabel('Attack Sucess (# of false negatives)' )
plt.show()
```
